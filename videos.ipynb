```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data = {
    'Name': ['Shumaila', 'Mehraj', 'Mehraj', 'Mehraj', 'Mehraj'],
    'Status': ['Done', 'Done', 'Done', 'Done', 'Done']
}
df = pd.DataFrame(data)
plt.figure(figsize=(10, 6))  
sns.countplot(x='Name', hue='Status', data=df)
plt.title('Video Status by Creator')
plt.xlabel('Creator')
plt.ylabel('Count')
plt.show()
```

```python
import pandas as pd
data = {
    'Name': ['Shumaila', 'Mehraj', 'Mehraj', 'Mehraj', 'Mehraj'],
    'Status': ['Done', 'Done', 'Done', 'Done', 'Done']
}
df = pd.DataFrame(data)
mehraj_videos = df[df['Name'] == 'Mehraj']
num_mehraj_videos = len(mehraj_videos)
shumaila_videos = df[df['Name'] == 'Shumaila']
num_shumaila_videos = len(shumaila_videos)
print(f"Mehraj made {num_mehraj_videos} videos.")
print(f"Shumaila made {num_shumaila_videos} videos.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
file_path = r'D:/pythonpro/Python Videos for Python Pro Channel (Morning) - Pytho Pro.csv'
df = pd.read_csv('C:/Users/SoftAsia Morning/Documents/pythonpro/Python Videos for Python Pro Channel (Morning) - Pytho Pro.csv')
mehraj_videos = df[df['Creator'] == 'Mehraj']
num_mehraj_videos = len(mehraj_videos)
shumaila_videos = df[df['Creator'] == 'Shumaila']
num_shumaila_videos = len(shumaila_videos)
rimsha_videos = df[df['Creator'] == 'Rimsha']
num_rimsha_videos = len(rimsha_videos)
creators = ['Mehraj', 'Shumaila', 'Rimsha']
video_counts = [num_mehraj_videos, num_shumaila_videos, num_rimsha_videos]
plt.figure(figsize=(10, 6))
plt.bar(creators, video_counts, color=['pink', 'red', 'yellow'])
plt.xlabel('Creator')
plt.ylabel('Number of Videos')
plt.title('Number of Videos by Creator')
plt.show()
```

```python
import pandas as pd
file_path = r'D:/pythonpro/Python Videos for Python Pro Channel (Morning) - Pytho Pro.csv'
df = pd.read_csv('C:/Users/SoftAsia Morning/Documents/pythonpro/Python Videos for Python Pro Channel (Morning) - Pytho Pro.csv')
mehraj_df = df[df['Creator'] == 'Mehraj']
shumaila_df = df[df['Creator'] == 'Shumaila']
rimsha_df = df[df['Creator'] == 'Rimsha']
total_time_mehraj = mehraj_df['Time (in minutes)'].sum()
total_time_shumaila = shumaila_df['Time (in minutes)'].sum()
total_time_rimsha = rimsha_df['Time (in minutes)'].sum()
print(f"Total time worked by Mehraj: {total_time_mehraj} minutes")
print(f"Total time worked by Shumaila: {total_time_shumaila} minutes")
print(f"Total time worked by Rimsha: {total_time_rimsha} minutes")
```
