```python
import pandas as pd

data = [

    [1, 'Bulbasaur', 'Grass', 'Poison', 318, 45, 49, 49, 65, 65, 45, 1, False],

    [2, 'Ivysaur', 'Grass', 'Poison', 405, 60, 62, 63, 80, 80, 60, 1, False],

    [3, 'Venusaur', 'Grass', 'Poison', 525, 80, 82, 83, 100, 100, 80, 1, False],

    [4, 'Charmander', 'Fire', '', 309, 39, 52, 43, 60, 50, 65, 1, False],

    [5, 'Charmeleon', 'Fire', '', 405, 58, 64, 58, 80, 65, 80, 1, False],



    [6, 'Squirtle', 'Water', '', 314, 44, 48, 65, 50, 64, 43, 1, False],

    [7, 'Wartortle', 'Water', '', 405, 59, 63, 80, 65, 80, 58, 1, False],

    [8, 'Blastoise', 'Water', '', 530, 79, 83, 100, 85, 105, 78, 1, False],

    [9, 'Caterpie', 'Bug', '', 195, 45, 30, 35, 20, 20, 45, 1, False],

    [10, 'Metapod', 'Bug', '', 205, 50, 20, 55, 25, 25, 30, 1, False],

   

]

columns = ['#', 'Name', 'Type1', 'Type2', 'Total', 'HP', 'Attack', 'Defense', 'SP.Atk', 'SP.Def', 'Speed', 'Generation', 'Legendary']

df = pd.DataFrame(data, columns=columns)

df.set_index('#', inplace=True)

df.reset_index(inplace=True)

df.to_csv('pokemon_large.csv', index=False)

print("CSV file 'pokemon_large.csv' saved successfully.")

```

```python
import pandas as pd
df = pd.read_csv('pokemon_large.csv')
print(df.head())
```

```python
import pandas as pd

import matplotlib.pyplot as plt

df = pd.read_csv('pokemon_large.csv')

type1_distribution = df['Type1'].value_counts()

type2_distribution = df['Type2'].value_counts()

all_types = type1_distribution.add(type2_distribution, fill_value=0)

plt.figure(figsize=(8, 8))

all_types.sort_values(ascending=False).plot(kind='pie', autopct='%1.1f%%', startangle=140)

plt.title('Distribution of Pokémon Types (All Types)')

plt.ylabel('')

plt.axis('equal')

plt.legend(title='Type', loc='upper right')

plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('pokemon_large.csv')
plt.figure(figsize=(10, 6))
plt.hist(df['Total'], bins=20, color='skyblue', edgecolor='black')  
plt.xlabel('Total')
plt.ylabel('Frequency')
plt.title('Distribution of Pokémon Totals')
plt.grid(axis='y')  
plt.show()
```

```python
import pandas as pd

import matplotlib.pyplot as plt

df = pd.read_csv('pokemon_large.csv')

plt.figure(figsize=(10, 6))

plt.boxplot(df['Total'], vert=False)  

plt.xlabel('Total')

plt.title('Box Plot of Pokémon Totals')

plt.grid(axis='x')  

plt.show()

```

```python
import pandas as pd

import matplotlib.pyplot as plt

df = pd.read_csv('pokemon_large.csv')

legendary_distribution = df['Legendary'].value_counts()

colors = ['black']  

plt.figure(figsize=(8, 8))

legendary_distribution.plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=colors)

plt.title('Distribution of Legendary Pokémon')

plt.ylabel('')

plt.axis('equal')

plt.legend(['Not Legendary', 'Legendary'], title='Legendary', loc='upper right')

plt.show()

```
